```python
'''get stopwords'''
import pandas as pd
#stopwords already saved from NLTK library in case of internet problem
stopwords=pd.read_csv('stopwords.csv')['sw']
swl=set(stopwords.tolist())
swl
```

```python
'''split'''
import pandas as pd
allSt=pd.read_csv('./tweet_emotions.csv')
allSt['content']=allSt['content'].str.replace(r'\@\S+\s','', regex=True).str.replace(r'[^A-Za-z\s]','', regex=True).str.lower().str.replace(r'im','', regex=True).apply(lambda x: [word for word in x.split() if word not in swl])
pf=pd.DataFrame({'word':allSt['content'].explode()}).join(allSt.drop('content',axis=1))
dic={}
for i in allSt["sentiment"].unique():
    dic[i]=pf[pf['sentiment']==i]
top={}
for sen,df in dic.items():
    top[sen]=df['word'].value_counts().head(10)
tops={i:j for i,j in sorted(top.items(),key=lambda x:x[0])}
#print(tops) found 'im' in many of the top words lists, but it have no exact emotions => filter it out in regex part
print(tops)
```
